```python
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

np.random.seed(1)
```

```python
x = np.random.uniform(0,10,size=(1000,))
```

```python
y = (0.25*x) + (np.random.normal(loc=0, scale=1, size=(1000,))) + (-1.25)
```

```python
plt.scatter(x, y,s=1)
```

```python
lr = LinearRegression()
lr.fit(x.reshape(-1,1), y)
y_pred = lr.predict(x.reshape(-1,1))
```

```python
plt.scatter(x, y,s=1)
plt.plot(x, y_pred, 'r')
```
